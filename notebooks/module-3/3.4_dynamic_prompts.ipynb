```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from dataclasses import dataclass
from langchain.agents.middleware import dynamic_prompt, ModelRequest

@dataclass
class LanguageContext:
    user_language: str = "Polish"

@dynamic_prompt
def user_language_prompt(request: ModelRequest) -> str:
    """Generate system prompt based on user role."""
    user_language = request.runtime.context.user_language
    if user_language is None:
        user_language = "Polish"

    base_prompt = "You are a helpful assistant."

    return f"{base_prompt} only respond in {user_language}."
```

```python
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

model = ChatOllama(model="lfm2.5-thinking")

agent = create_agent(
    model=model,
    context_schema=LanguageContext,
    middleware=[user_language_prompt]
)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"message": [HumanMessage(content="Jak siÄ™ masz?")]},
    context=LanguageContext(user_language=None)
)

print(response["messages"][-1].content)
```
